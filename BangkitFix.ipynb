```python
#Load Library
from sklearn.cluster import KMeans, DBSCAN
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("/content/drive/MyDrive/datasetbansos.csv")
display(dataset)
```

```python
dataset = dataset.rename(columns={'No': 'no', 'Provinsi': 'provinsi',
                        'Kab/Kota': 'kab_kota',
                        'Kecamatan': 'kec','Desa':'desa','NIK':'nik','Nama':'nama','Penghasilan/Bulan':'gaji',
                        'Jumlah Tanggung Jawab':'j_tanggungjawab','Umur':'umur','Ibu Hamil':'ibu_hamil',
                        'Lanjut Usia':'lanjut_usia','Disabilitas':'disabilitas','Jumlah Motor':'j_motor',
                        'Jumlah Mobil':'j_mobil','BPNT':'bpnt','BST':'bst','PKH':'pkh','Layak Menerima Bantuan?':'layak_menerima'})
 
display(dataset)
```

```python
Features = dataset.drop(["provinsi","kab_kota","kec","desa","nama","no","umur","nik","ibu_hamil","lanjut_usia","disabilitas","bpnt","bst","pkh"], axis = 1).columns
display(Features)
```

```python
dataset['layak_menerima'].replace(['TIDAK', 'YA'], [0,1], inplace=True)
display(dataset.head())
```

```python
dataset=dataset[[
    "gaji",
    "j_tanggungjawab",
    "j_motor",
    "j_mobil",
    "layak_menerima"
]].dropna(axis=0, how='any')
```

```python
X = dataset[['gaji', 'j_tanggungjawab','j_motor','j_mobil']].values
y = dataset['layak_menerima'].values
print('-'*80)
print(f'Shape of X is {X.shape}\nShape of y is {y.shape}')

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
print('-'*80)
print(f"Lenght of X_train: {len(X_train)}\nLenght of X_test: {len(X_test)}")
print(f"Lenght of y_train: {len(y_train)}\nLenght of y_test: {len(y_test)}")
```

```python
# Fitting Naive Bayes to the Training set
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
print(model)
model.fit(X_train, y_train)

# Predicting the Test set results
y_pred = model.predict(X_test)

# Accuracy on the Test set results                              
from sklearn.metrics import accuracy_score
print('\n'+'-'*20+'Accuracy Score on the Test set'+'-'*20)                             
print("{:.0%}".format(accuracy_score(y_test,y_pred)))
```
